```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/home/naka/code/egfr_binder_rd2/src/egfr_binder_rd2/data/melted_residue_distances.csv')
```

```python
idx = df['set'].str.contains('egf')
df.loc[idx, 'binder'] = 'egf'

idx = df['set'].str.contains('panitumumab')
df.loc[idx, 'binder'] = 'panitumumab'

idx = df['set'].str.contains('cetuximab')
df.loc[idx, 'binder'] = 'cetuximab'
```

```python
f,ax = plt.subplots(figsize=(10,5))
sns.scatterplot(data=df, x='adjusted_position', y='distance', hue='binder', ax=ax)
```
